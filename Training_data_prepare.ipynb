#### Formatting the data for rasa

```
# Test data
# Function to format the training data
def prepare_rasa_training_data(raw_data_list, label_name, make_sentence_list):
  nlu_md = """
## intent:inform"""
  # raw_data_list = map(str.lower, raw_data_list)
  random.shuffle(raw_data_list)
  for data in raw_data_list:
    for append_string in make_sentence_list:
      entity = "[%s](%s)" % (data, label_name)
      data_item = append_string % entity
      train_data = "\n- %s" % data_item
      nlu_md = nlu_md + train_data
  return nlu_md
```
